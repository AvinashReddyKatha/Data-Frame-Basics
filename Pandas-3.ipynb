```python
import pandas as pd
```

```python
weather=pd.DataFrame({
    "City":["Hyderabad","Mumbai","Delhi","Banglore","Chennai"],
    "Tempearature":[25,45,35,26,45],
    "Humidity":[60,56,80,67,68],
    
})
weather
```

```python
weather2=pd.DataFrame({
    "City":["Hyderabad","UP","Bhopal","Kerala","Vizag"],
    "Tempearature":[25,45,35,26,45],
    "Humidity":[60,56,80,67,68],
    
})
weather2
```

```python
df=pd.concat([weather,weather2])
df
```

```python
#If we need continous data then just add ignore_index=True
#it will give us continous sereis of table

df1=pd.concat([weather,weather2],ignore_index=True)
df1
```

```python
df1=pd.concat([weather,weather2],keys=["weather","weather2"])
df1
#This keys is used to seperate and show the data individually of our diffrent dataframes
```

```python
df1.loc["weather"]
```

```python
tempdata=pd.DataFrame({
    "city":["Mumbai","Delhi","Chennai"],
    "temperature":[26,25,36],
    
},index=[0,1,2])
tempdata
```

```python
winddata=pd.DataFrame({
    "city":["Mumbai","Delhi"],
    "windspeed":[26,25]
},index=[0,1])
winddata
#index=[0,1] will be used to give order when we are performing concate functunction
```

```python
df=pd.concat([tempdata,winddata],axis=1)
df
```

```python
#Here we can concate the series data for tha any table
```
